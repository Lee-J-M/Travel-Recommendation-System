```python
#후기 내용만 담은 txt 파일을 하나의 리스트에 저장
#리스트에 저장된 파일들은 한 나라에 대한 문서
file_arr=['다낭호이안1.txt','다낭호이안2.txt','다낭호이안3.txt','다낭호이안4.txt','다낭호이안5.txt','다낭호이안6.txt']

line_arr=[] #txt를 문자열로 받음
for i in range(len(file_arr)):
    file=open(file_arr[i],'r',encoding='utf-8')
    line_arr.append(file.read())
    file.close
# print(line_arr)
```

```python
# 형용사만 추출하는 함수
from konlpy.tag import Okt
okt = Okt()

def tokenize(text):
    stem=[]
    tagged = okt.pos(text)
    for i in range(0,len(tagged)):
        if(tagged[i][1]=='Adjective'):
            stem.append(tagged[i][0])     
    return stem
```

```python
tokenized_word=[] #찢어진 단어
# for i in range(len(line_arr)):
#     tokenized_word.append(tokenize(line_arr[i]))
for i in line_arr:
    tokenized_word.append(tokenize(i))
# print(tokenized_word)
```

```python
# 단순히 단어의 빈도수를 세는 함수
def f(t, d):
    # t는 토큰한 단어, d는 문서
    return d.count(t)

max_word=[]#문서별 가장 많이 나온 단어 수 

# max_word를 초기화
for i in range(len(line_arr)):
    max_word.append(0)

#max단어수를 계산
for i in range(len(tokenized_word)):
    for j in tokenized_word[i]:
        if(max_word[i]<f(j,line_arr[i])):
            max_word[i]=f(j,line_arr[i])
# print(max_word)

def tf(t,d,i):
    # t는 토큰화 된 단어,d 는 문서,i는 인덱스
    return 0.5 + 0.5*f(t,d)/max_word[i]
```

```python
# 문서별 tf값 확인
for i in range(len(tokenized_word)):
    print('===========================',i+1,'번째 문서===============================')#문서가 바뀜을 표현
    for j in tokenized_word[i]:
        print(tf(j,line_arr[i],i),j)
```
